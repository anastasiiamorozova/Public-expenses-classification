```python
import pandas as pd
```

```python
path = 'zbiorcze1.csv'
```

```python
df = pd.read_csv(path)
df
```

```python
df['hashtags'] = ''
```

```python
hashtags_keywords = {
    'Straż pożarna': ['straż','pożar', 'osp', 'ochotnicz','psp','państwowa straż pożarna', 'komenda powiatowej państwowej straży pożarnej', 'komendy powiatowej państwowej straży pożarnej'],
    'Kościół': ['bractw', 'święt','parafia', 'caritas', 'diecez', 'archidiecez', 'seminarium', 'kościół', 'opactw', 'kuri', 'klasztor', 'zakon', 'monaster'],
    'Szpital': ['szpital', 'rehabilit', 'terapi'],
    'Koła Gospodyń Wiejskich': ['koło gospodyń wiejskich'],
    'Policja': ['policj', 'policjantów', 'mswia', 'komend'],
    'Organizacje leśne': ['leśn','las','ptl', 'santa', 'ekolog','rolp', 'przyrod', 'lop', 'sitlid', 'stowarzyszenie inżynierów','zespół sygnalistów', 'przyjaźń', 'szkół leśnych','santa', 'scholae', 'fundacja dla niepodległej'],
    'Szkoły': ['szkół', 'szko', 'uniwersytet', 'szkoł', 'uczniow','uczniów' ,'technikum', 'sggw', 'oświat'],
    'Myśliwi': ['myśliw', 'akteon', 'łowieck', 'strzel'],
    'Sport i kultura': ['sport','rekreac', 'kultur']}
```

```python
for index, row in df.iterrows():
    company_name = str(row['NAZWA FUNDACJI']).lower()
    hashtags = [hashtag for hashtag, keywords in hashtags_keywords.items() if any(keyword in company_name for keyword in keywords)]
    df.at[index, 'hashtag'] = ', '.join(hashtags)
```

```python
df = df.sort_values(by='hashtag', ascending=False)
```
